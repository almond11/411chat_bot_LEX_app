```python
import pandas as pd
```

```python
metadata = pd.read_csv('recipe.csv')
```

```python
l = len(metadata)
print(l)
```

```python
# create new feature 'diet'

for i in range(l):
    if metadata.iloc[i][4] <= 200:
        metadata.at[i, 'diet'] = 'light'
    elif metadata.iloc[i][4] <= 600:
        metadata.at[i, 'diet'] = 'medium'
    elif metadata.iloc[i][4] > 600:
        metadata.at[i, 'diet'] = 'heavy'
```

```python
# create new feature 'low-fat'

for i in range(l):
    if metadata.iloc[i][5] <= 20:
        metadata.at[i, 'low-fat'] = 'low-fat'
```

```python
# create new feature 'high-protein'

for i in range(l):
    if metadata.iloc[i][6] >= 30:
        metadata.at[i, 'high-protein'] = 'high-protein'
```

```python
# create new feature 'low-sodium'

for i in range(l):
    if metadata.iloc[i][7] <= 50:
        metadata.at[i, 'low-sodium'] = 'low-sodium'
```

```python
df = metadata[['RecipeId', 'title', 'ingredient', 'rating', 'diet', 'low-fat', 'high-protein', 'low-sodium']]
```

```python
df.at[0,'ingredients'] = df.iloc[0]['ingredient']
for i in range(1, l):
    if df.iloc[i]['title'] == df.iloc[i-1]['title']:
        df.at[i,'ingredients'] = df.iloc[i-1]['ingredients'] + ' '+ df.iloc[i]['ingredient']
    else: 
        df.at[i,'ingredients'] = df.iloc[i]['ingredient']
```

```python
for i in range(l2):
    df4.at[i,'bag'] = df4.iloc[i]['ingredients']
    if df4.iloc[i]['diet'] in ['light', 'medium', 'heavy']:
        df4.at[i,'bag'] = df4.iloc[i]['bag'] + ' ' + df4.iloc[i]['diet']
        if df4.iloc[i]['low-fat']=='low-fat':
            df4.at[i,'bag'] = df4.at[i,'bag'] + ' ' + df4.iloc[i]['low-fat']
            if df4.iloc[i]['high-protein']=='high-protein':
                df4.at[i,'bag'] = df4.at[i,'bag'] + ' ' + df4.iloc[i]['high-protein']
                if df4.iloc[i]['low-sodium']=='low-sodium':
                    df4.at[i,'bag'] = df4.at[i,'bag'] + ' ' + df4.iloc[i]['low-sodium']
```

```python
## Choose only RecipleId column and bag column

df5 = df4[['RecipeId', 'bag']]
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# instantiating and generating the count matrix
count = CountVectorizer()
count_matrix = count.fit_transform(df5['bag'])

# generating the cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
cosine_sim
```

```python
# creating a Series for the recipe titles so they are associated to an ordered numerical list 
# I will use in the function to match the indexes

indices = pd.Series(df5.index)
indices.head(3)
```

```python
#  defining the function that takes in recipe title as input and returns the top 10 recommended recipes
def recommendations(title, cosine_sim = cosine_sim):
    
    # initializing the empty list of recommended recipes
    recommended_recipes = []
    
    # getting the index of the recipe that matches the title
    idx = indices[indices == title].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar recipes
    top_10_indexes = list(score_series.iloc[1:11].index)
    
    # populating the list with the titles of the best 10 matching recipes
    for i in top_10_indexes:
        recommended_recipes.append(list(df5.index)[i])
        
    return recommended_recipes
```

```python
# Spot check step 1

recommendations(0)
```

```python
# Spot check step 2

count = 1
for i in recommendations(0):
    print(count, df5.iloc[i][0])
    print(df5.iloc[i][1])
    count += 1
```

```python
df6 = pd.DataFrame(columns=['recipe', 'recommend'])
print(df6)
```

```python
for i in range(l2):
    df_temp = pd.DataFrame([[i,recommendations(i)]], columns=['recipe', 'recommend'])
    df6 = pd.concat([df6, df_temp])
```
